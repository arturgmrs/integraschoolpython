Aplicação para extrair dados de páginas web usando [Tabula](https://github.com/chezou/tabula-py)  

```
# instalar o pacote tabula no python
pip install tabula-py
```

```
import tabula

tabula.environment_info()
```

```
# fonte do pdf
pdf_path = "http://www.comperve.ufrn.br/conteudo/metropoledigital/2020/documentos/lista_aprovados_terceira_chamada_20200131.pdf"

# Opções de configuração da página para leitura, no exemplo é lida a página 3 do edital
tabula.read_pdf(pdf_path, pages=3, stream=True)[0]
```
